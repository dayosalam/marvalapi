```python
# Built a function that passes the information of a character that i needed into a list 



def descrip(char_name):
    
    Id = []
    name = []
    comics = []
    series = []
    stories = []
    events = []
    
    
    my_char = characters.all(name=char_name)["data"]["results"]
    for char in my_char:
        Id = Id + [char["id"]]
        for char in my_char:
            name = name + [char["name"]]
        for comic in char["comics"]["items"]:
            comics.append(comic["name"])
            for char in my_char:
                name = name + [char["name"]]
        #     print("-" * 50)
        for serie in char["series"]["items"]:
            series.append(serie["name"])
        #     #     print("-" * 50)
        for story in char["stories"]["items"]:
            stories.append(story["name"])
        #     #     print("-" * 50)
        for event in char["events"]["items"]:
            events.append(event["name"])

    return Id, name, comics, series, stories, events

```

```python
# Created a new list then passed the list created in the function into the new list then passed that new list into a single dictionary
n_Id = []
n_name = []
n_comics = []
n_series = []
n_stories = []
n_events = []
for char_name in marvel_characters_list:
    Id, name, comics, series, stories, events = descrip(char_name)
    n_Id = n_Id + Id
    n_name = n_name + name
    n_comics = n_comics + comics
    n_series = n_series + series
    n_stories = n_stories + stories
    n_events = n_events + events
a_dict = {}
a_dict["ID"] = n_Id
a_dict["Character Name"] = n_name
a_dict["Comics"] = n_comics
a_dict["Series"] = n_series
a_dict["Story"] = n_stories
a_dict["Events"] = n_events
```

```python
# Built a function to check which characters did not pass into the new list because of different issues like spellings or the character was not even in the APIs database
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value not in lst2]
    return lst3


```
